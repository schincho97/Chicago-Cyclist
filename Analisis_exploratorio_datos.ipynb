# **Primer Trimestre 2018**

```python
Q12018 = pd.read_csv('Divvy_Trips_2018_Q1.csv')
Q12018.head()
```

```python
# Poner los mismos nombres en las columnas ya que el primer trimestre viene cambiado el nombre de las columnas.
columns= ['trip_id','start_time','end_time','bikeid','tripduration','from_station_id','from_station_name','to_station_id','to_station_name','usertype','gender','birthyear']
```

```python
Q12018.columns = columns
```

```python
Q12018.head()
```

```python
# Guardar los dataset limpios en un archivo CSV
Q12018.to_csv('Q12018.csv', index=False)
```

```python
#Columna con edad a año 2019 de los clientes
Q12018['Age'] = 2019 - Q12018['birthyear']
```

```python
Q12018.head()
```

```python
# ver extension del dataset
Q12018.shape
```

```python
Q12018.info()
```

```python
# Convertir la columna de fecha y hora a tipo datetime
Q12018['start_time'] = pd.to_datetime(Q12018['start_time'])
Q12018['end_time'] = pd.to_datetime(Q12018['end_time'])
```

```python
# Extraer componentes de fecha y hora en columnas separadas
Q12018['start_day'] = Q12018['start_time'].dt.day
Q12018['start_month'] = Q12018['start_time'].dt.month
Q12018['start_hour'] = Q12018['start_time'].dt.hour
Q12018['start_year'] = Q12018['start_time'].dt.year
Q12018['end_day'] = Q12018['end_time'].dt.day
Q12018['end_month'] = Q12018['end_time'].dt.month
Q12018['end_hour'] = Q12018['end_time'].dt.hour
Q12018['end_year'] = Q12018['end_time'].dt.year
```

```python
# Eliminar columnas no hacen falta
columnas_a_eliminar = ['trip_id', 'start_time', 'end_time','start_hour','end_hour','bikeid', 'from_station_id', 'to_station_id','end_day','end_month','end_year']
Q1_2018 = Q12018.drop(columnas_a_eliminar, axis=1)
```

```python
Q1_2018.head()
```

```python
Q1_2018.describe()
```

```python
#Calculo de valores faltantes
valores_faltantes = Q1_2018.isnull().sum()
valores_faltantes
```

```python
# Ver que proporcion tienen conforme el total los valores faltantes
proporcion_faltantes = valores_faltantes / len(Q1_2018) * 100 #proporcion de valores faltantes
proporcion_faltantes
```

TRABAJAR COLUMNA GENDER

```python
# Calcular la moda
moda = Q1_2018['gender'].mode()[0]

# Rellenar la columna con la moda
Q1_2018['gender'].fillna(moda, inplace=True)
```

```python
Q1_2018['gender'].value_counts()
```

```python
sns.countplot(data=Q1_2018, x='gender')
```

TRABAJAR CON COLUMNA BIRTHYEAR

```python
# Calcular la mediana
mediana = Q1_2018['birthyear'].median()

# Rellenar la columna con la mediana
Q1_2018['birthyear'].fillna(mediana, inplace=True)
Q1_2018['birthyear'].value_counts()
```

```python
Q1_2018['birthyear'].value_counts()
```

TRABAJAR CON COLUMNA AGE

```python
# Calcular la mediana
mediana = Q1_2018['Age'].median()

# Rellenar la columna con la mediana
Q1_2018['Age'].fillna(mediana, inplace=True)
```

```python
sns.boxplot(data=Q1_2018, x='Age')
```

```python
# Calcular el rango intercuartílico
q1 = Q1_2018['Age'].quantile(0.25)
q3 = Q1_2018['Age'].quantile(0.75)
rango_intercuartilico = q3 - q1

# Imprimir el rango intercuartílico
print("Rango intercuartílico:", rango_intercuartilico)
```

```python
# Definir los rangos de edad
Age_range = [(0, 17), (18, 29), (30, 50), (50, 65), (66, 100)]

# Asignar el rango de edad correspondiente a cada registro
def asignar_rango_edad(Age):
    for rango in Age_range:
        if rango[0] <= Age <= rango[1]:
            return f"{rango[0]}-{rango[1]}"
    return "100+"

Q1_2018['Age_range'] = Q1_2018['Age'].apply(asignar_rango_edad)

```

```python
Q1_2018.head()
```

```python
Q1_2018['Age_range'].value_counts()
```

```python
sns.countplot(data=Q1_2018, x='Age_range')
```

```python
# Filtrar los registros con rangos de edad mayores y menores
edad_menor = 18
edad_mayor = 65

# Crear una máscara booleana
mascara = Q1_2018['Age_range'].apply(lambda x: True if '-' in x else False)
mascara &= Q1_2018['Age_range'].apply(lambda x: int(x.split('-')[0]) if '-' in x else int(x[:-1])) >= edad_menor
mascara &= Q1_2018['Age_range'].apply(lambda x: int(x.split('-')[1]) if '-' in x else int(x[:-1])) <= edad_mayor
```

```python
mascara.value_counts()
```

```python
# Filtrar el dataset utilizando la máscara booleana
Q1_Edad = Q1_2018[mascara]
Q1_Edad.head()
```

```python
Q1_Edad.describe()
```

TRABAJAR CON TRIPDURATION

```python
# Eliminar las comas de la columna 'tripduration'
Q1_Edad.loc[:, 'tripduration'] = Q1_Edad['tripduration'].str.replace(',', '')
# Convertir la columna a tipo numérico
Q1_Edad.loc[:, 'tripduration'] = pd.to_numeric(Q1_Edad['tripduration'], errors='coerce')
# Calcular el percentil 99.9
percentil_999 = Q1_Edad['tripduration'].quantile(0.999)
```

```python
# Filtrar los valores al percentil 99.9
Q1_TRIP = Q1_Edad[Q1_Edad['tripduration'] <= percentil_999]
```

```python
Q1_TRIP.head()
```

```python
Q1_TRIP['tripduration'].describe()
```

```python
# Calcular el rango intercuartílico
q1 = Q1_TRIP['tripduration'].quantile(0.25)
q3 = Q1_TRIP['tripduration'].quantile(0.75)
rango_intercuartilico = q3 - q1

# Imprimir el rango intercuartílico
print("Rango intercuartílico:", rango_intercuartilico)
```

Ver distribucion de todos los features

```python
# Obtener la lista de nombres de columnas
columnas = ['tripduration','Age','start_day']
# Generar los boxplots para cada columna
for columna in columnas:
   if Q1_TRIP[columna].dtype != 'object':
        sns.catplot(data=Q1_TRIP, x=columna, kind='box')
        plt.title(f'Boxplot de {columna}')
        plt.xlabel(columna)
        plt.ylabel('Valor')
        plt.show()
```

```python
# Ver si esta balanceado
usuarios = Q1_2018['usertype'].value_counts()
usuarios
```

```python
sns.countplot(data=Q12018, x='usertype')
```

Guardar dos datasets limpios de subscribe y customer

```python
# Definir los rangos de edad
Age_range = [(0, 17), (18, 29), (30, 50), (50, 65), (66, 100)]

# Asignar el rango de edad correspondiente a cada registro
def asignar_rango_edad(Age):
    for rango in Age_range:
        if rango[0] <= Age <= rango[1]:
            return f"{rango[0]}-{rango[1]}"
    return "100+"

Q4_2019['Age_range'] = Q4_2019['Age'].apply(asignar_rango_edad)
```
